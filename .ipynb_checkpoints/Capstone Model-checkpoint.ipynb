```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split

from matplotlib.image import imread
from PIL import Image
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Input
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint
import tensorflow
import numpy.random
from numpy.random import seed
seed(1)
tensorflow.random.set_seed(seed = 958)
from tensorflow.keras.layers import BatchNormalization
from sklearn.preprocessing import minmax_scale
from sklearn.preprocessing import MinMaxScaler
import urllib.request
```

```python
data = ImageDataGenerator(rescale = 1./255, validation_split=0.3, 
                             rotation_range = 10,
                             zoom_range=[0.9, 1.1]
                                )
```

```python
training = data.flow_from_directory(directory = "/Users/sunnychahal/Capstone Data",
                                   target_size = (128,128),
                                   color_mode='rgb',                                     
                                   class_mode = "binary",
                                   batch_size = 32,
                                   subset='training')
```

```python
validation = data.flow_from_directory(directory = "/Users/sunnychahal/Capstone Data",
                                   target_size = (128,128),
                                   color_mode='rgb',
                                   class_mode = "binary",
                                   batch_size = 32,
                                   subset='validation')
```

```python
training[0][1][15]
```

```python
plt.imshow(training[0][0][15])
```

```python
cnn_model = Sequential()
cnn_model.add(Input(shape=(128,128,3)))

cnn_model.add(Conv2D(filters = 96,
                      kernel_size = 3,
                       activation='relu'))
#cnn_model.add(MaxPooling2D(pool_size=(2,2)))
#cnn_model.add(Conv2D(filters = 72,
#                       kernel_size = 3,
#                       activation='relu'))
#cnn_model.add(MaxPooling2D(pool_size=(2,2)))
cnn_model.add(Conv2D(filters = 64,
                       kernel_size = 3,
                       activation='relu'))
cnn_model.add(MaxPooling2D(pool_size=(2,2)))
cnn_model.add(BatchNormalization())
cnn_model.add(Flatten())
cnn_model.add(Dense(1024, activation='relu'))
#cnn_model.add(Dropout(0.15))
cnn_model.add(Dense(128, activation='relu'))
cnn_model.add(Dropout(0.15))
#cnn_model.add(Dense(32, activation='relu'))
#cnn_model.add(Dropout(0.15))
#cnn_model.add(Dense(768, activation='relu'))
#cnn_model.add(Dropout(0.15))
#cnn_model.add(Dense(128, activation='relu'))
#cnn_model.add(Dropout(0.5))

cnn_model.add(Dense(2, activation='sigmoid'))
```

```python
#cnn_model.summary()
```

```python
SGD = tensorflow.keras.optimizers.SGD(learning_rate=0.0001, momentum=0, nesterov=False)
```

```python
cnn_model.compile(loss='categorical_crossentropy',
                  optimizer= SGD,
                  metrics=['accuracy'])
```

```python
#filepath="bestmodel.h5"
#checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
#callbacks_list = [checkpoint]
```

```python
cnn_model.fit(training, validation_data= validation, epochs = 50, verbose = 1)
```

```python
cnn_score = cnn_model.evaluate(validation, verbose=1)
```
